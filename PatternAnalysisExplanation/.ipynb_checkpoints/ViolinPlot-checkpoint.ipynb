```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from rdflib import Graph
from rdflib.namespace import RDF
import re
import sys

from rdflib.plugins.sparql.processor import SPARQLResult

def sparql_results_to_df(results: SPARQLResult) -> pd.DataFrame:
    """
    Export results from an rdflib SPARQL query into a `pandas.DataFrame`,
    using Python types. See https://github.com/RDFLib/rdflib/issues/1179.
    """
    return pd.DataFrame(
        data=([None if x is None else x.toPython() for x in row] for row in results),
        columns=[str(x) for x in results.vars],
    )
```

```python
g1 = Graph()
g1.parse("../../store_data/ImProVIT/All_Measurement_parameters_D0_HBsRE_v5_1.nt", format="nt")
len(g1)
```

```python
query = """
prefix improvit: <http://www.project-improvit.de/vocab/> 
select ?s (count(?o) as ?NumOc)
                where {
                {?s ?p ?o}
                UNION
                {?o ?p ?s}
                }
                GROUP BY ?s
"""
qres = g1.query(query)
entity = sparql_results_to_df(qres)
entity
```

```python
entity = entity.loc[entity.NumOc>1]
entity = entity.dropna(subset=['s'])
entity = entity[entity.s.str.contains('http:', na=False)]
entity
```

```python
query = """
prefix improvit: <http://www.project-improvit.de/vocab/> 
select distinct ?p
       where {
                ?s ?p ?o
                }
                
"""
qres = g1.query(query)
relation = sparql_results_to_df(qres)
relation
```

# Box Plots

```python
col_study = ['CD3pos_CD8pos_CD14neg_CD19neg', 'CD3pos_CD8pos_CD14neg_CD19neg_CD45RAneg_CCR7neg',
             'CD3pos_CD8pos_CD14neg_CD19neg_CD45RAneg_CCR7pos', 'CD3pos_CD8pos_CD14neg_CD19neg_HLA_DRpos',
             'CD3pos_CD8pos_CD14neg_CD19neg_Ki67pos']

def data_boxplot(data_kmeans, col_study, label):
    data_cytometric = pd.DataFrame(columns=['parameter', 'result'])
    for parameter in col_study:
        df = pd.DataFrame(columns=['parameter', 'result'])
        df['result'] = data_kmeans[(data_kmeans[parameter]!=-1000) & (data_kmeans['response']==label) &
                                   (data_kmeans['cluster']==0)][parameter].values
        df['parameter'] = 'Community1-' + parameter
        data_cytometric = pd.concat([data_cytometric, df])
        df = pd.DataFrame(columns=['parameter', 'result'])
        df['result'] = data_kmeans[(data_kmeans[parameter]!=-1000) & (data_kmeans['response']==label) &
                                   (data_kmeans['cluster']==1)][parameter].values
        df['parameter'] = 'Community2-' + parameter
        data_cytometric = pd.concat([data_cytometric, df])
    return data_cytometric
```

```python
data_cytometric = data_boxplot(data_kmeans, col_study, 'cured')
fig, ax = plt.subplots(figsize=(6, 3))
# Plot the orbital period with horizontal boxes
sns.boxplot(x="result", y="parameter", data=data_cytometric,
            whis=[0, 100], width=.5, palette="vlag")

# Add in points to show each observation
sns.stripplot(x="result", y="parameter", data=data_cytometric,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
ax.set(xlabel="")
sns.despine(left=True)
# plt.savefig('Boxplot/RESCAL_SemEP_cured.pdf', bbox_inches='tight', format='pdf', transparent=True)
# plt.savefig('Boxplot/RESCAL_METIS_cured.pdf', bbox_inches='tight', format='pdf', transparent=True)
plt.savefig('Boxplot/RESCAL_KMeans_cured.pdf', bbox_inches='tight', format='pdf', transparent=True)
```

```python
data_cytometric = data_boxplot(data_kmeans, col_study, 'non_cured')
fig, ax = plt.subplots(figsize=(5, 3))
# Plot the orbital period with horizontal boxes
sns.boxplot(x="result", y="parameter", data=data_cytometric,
            whis=[0, 100], width=.5, palette="vlag")

# Add in points to show each observation
sns.stripplot(x="result", y="parameter", data=data_cytometric,
              size=4, color=".3", linewidth=0)

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
ax.set(xlabel="")
sns.despine(left=True)
# plt.savefig('Boxplot/RESCAL_SemEP_noncured.pdf', bbox_inches='tight', format='pdf', transparent=True)
# plt.savefig('Boxplot/RESCAL_METIS_noncured.pdf', bbox_inches='tight', format='pdf', transparent=True)
plt.savefig('Boxplot/RESCAL_KMeans_noncured.pdf', bbox_inches='tight', format='pdf', transparent=True)
```

```python
# col_study = ['CD3pos_CD8pos_CD14neg_CD19neg', 'CD3pos_CD8pos_CD14neg_CD19neg_CD45RAneg_CCR7neg',
#              'CD3pos_CD8pos_CD14neg_CD19neg_CD45RAneg_CCR7pos', 'CD3pos_CD8pos_CD14neg_CD19neg_HLA_DRpos',
#              'CD3pos_CD8pos_CD14neg_CD19neg_Ki67pos']
replacement_mapping_dict = {'CD3pos_CD8pos_CD14neg_CD19neg': 'CD8+ T cells',
                            'CD3pos_CD8pos_CD14neg_CD19neg_CD45RAneg_CCR7neg': 'CD8+ TEM cells',
                           'CD3pos_CD8pos_CD14neg_CD19neg_CD45RAneg_CCR7pos': 'CD8+ TCM cells',
                           'CD3pos_CD8pos_CD14neg_CD19neg_HLA_DRpos': 'CD8+ HLA_DR+ T cells',
                           'CD3pos_CD8pos_CD14neg_CD19neg_Ki67pos': 'CD8+ Ki67+ T cells',}
#                            'CD3pos_CD8pos_CD14neg_CD19neg_PD1pos': 'CD8+ PD1+ T cells'}
def data_violinplot(data_kmeans, replacement_mapping_dict, cls):
    data_cytometric = pd.DataFrame(columns=['parameter', 'Parameter values', 'Functional HBV'])
    for key, val in replacement_mapping_dict.items():
        df = pd.DataFrame(columns=['parameter', 'Parameter values', 'Functional HBV'])
        df['Parameter values'] = data_kmeans[(data_kmeans[key]!=-1000) & (data_kmeans['response']=='cured') &
                                   (data_kmeans['cluster']==cls)][key].values
        df['parameter'] = val
        df['Functional HBV'] = 'cured'
        data_cytometric = pd.concat([data_cytometric, df])
        df = pd.DataFrame(columns=['parameter', 'result', 'Hepatitis B Cure'])
        df['Parameter values'] = data_kmeans[(data_kmeans[key]!=-1000) & (data_kmeans['response']=='non_cured') &
                                   (data_kmeans['cluster']==cls)][key].values
        df['parameter'] = val
        df['Functional HBV'] = 'non_cured'
        data_cytometric = pd.concat([data_cytometric, df])
#     data_cytometric['parameter'].replace(replacement_mapping_dict, inplace=True)
    return data_cytometric
```

```python
data_cytometric = data_violinplot(data_kmeans, replacement_mapping_dict, 0)
fig, ax = plt.subplots(figsize=(10, 4))
sns.violinplot(data=data_cytometric, x="parameter", y="Parameter values", hue="Functional HBV")#, inner="stick"
ax.set(xlabel="")
plt.ylim(-5, 78)
# plt.xticks(fontsize=10, rotation='vertical')
# plt.savefig('Boxplot/RESCAL_SemEP_C1.pdf', bbox_inches='tight', format='pdf', transparent=True)
# plt.savefig('Boxplot/RESCAL_METIS_C1.pdf', bbox_inches='tight', format='pdf', transparent=True)
plt.savefig('Boxplot/RESCAL_KMeans_C1.pdf', bbox_inches='tight', format='pdf', transparent=True)
```
