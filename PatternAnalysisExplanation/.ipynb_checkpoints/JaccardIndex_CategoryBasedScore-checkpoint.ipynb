## Get standards

```python
qres = g.query(
    """PREFIX owl: <http://www.w3.org/2002/07/owl#>
       PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
       PREFIX sto: <https://w3id.org/i40/sto#>
    
    select distinct ?s where {
            ?s rdf:type sto:Standard .
    }""")
list_sto = []
for row in qres:
    list_sto.append(str("%s" %row['s']))

list_sto = [s.replace('https://w3id.org/i40/sto#', '') for s in list_sto]
print(list_sto, len(list_sto))
```

## Get relatedTo property

```python
#sto_related = pd.DataFrame(columns=['sto1', 'sto2'])
qres = g.query(
    """PREFIX owl: <http://www.w3.org/2002/07/owl#>
       PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
       PREFIX sto: <https://w3id.org/i40/sto#>
    
    select distinct ?s1 ?s2 where {
            ?s1 rdf:type sto:Standard .
            ?s1 sto:relatedTo ?s2 .
    }""")
    

sto_related = pd.DataFrame(qres.bindings)
sto_related.columns = ['s1', 's2']
sto_related.replace('https://w3id.org/i40/sto#', '', regex=True, inplace=True)
(sto_related)
```

## Get standards with property RelatedTo

```python
list_sto = set(sto_related.s1.values)
list_sto.update(sto_related.s2.values)

print(list_sto, len(list_sto))
```

## Datalog Program

```python
#sto_related.iloc[4].values[1]
#sto_related.loc[sto_related.index[0], 's1']
from pyDatalog import pyDatalog
from pyDatalog.pyDatalog import assert_fact, load, ask

pyDatalog.create_terms('relatedTO, same_equivalent_class, X, Y, Z')
# Extensiona Database
for sto in sto_related.values:
    assert_fact('relatedTO', sto[0], sto[1])

# Reflexive property
#same_equivalent_class(X,X) <= relatedTO(X,_)
# Symmetric property
#same_equivalent_class(Y,X) <= relatedTO(X,Y)

# Symmetric property
relatedTO(X,Y) <= relatedTO(Y,X)
# Base case
same_equivalent_class(X,Y) <= relatedTO(X,Y)
# Inductive case
same_equivalent_class(X,Y) <= relatedTO(X,Z) & same_equivalent_class(Z,Y) #& (X!=Y)  # via Z
```

## Compute Equivalente Class

```python
equivalence_class = []
for sto in list_sto:
    a = same_equivalent_class(sto, X)
    print('result inductive case')
    flag = True
    for ec in equivalence_class:
        if (set(a).issubset(ec)):   #a <= equivalence_class
            flag = False
            break
        if set(ec).issubset(a):
            equivalence_class.remove(ec)
            flag = True
            break
    if flag:
        equivalence_class.append(a)
```

```python
import os
from os.path import isfile, join
from os import listdir
import pandas as pd
```

```python
#===========  Euclidean  ================
#cls_measure = 'clusteringMeasures/SemEPTransH'   #'../../data/Input/'
#cls_measure = 'clusteringMeasures/KMeansTransH'

#===========  Cosine  ================
#cls_measure = 'Result_by_sim_measure/Cosine/Cluster/SemEPTransH'
#cls_measure = 'Result_by_sim_measure/Cosine/Cluster/METISTransH'
#cls_measure = 'Result_by_sim_measure/Cosine/Cluster/KMeansTransH'

cls_measure = 'Result_by_sim_measure/Euclidean/kmeans_compute_k/clusteringMeasures/SemEPTransH'
#cls_measure = 'Result_by_sim_measure/Euclidean/kmeans_compute_k/clusteringMeasures/METISTransH'
#cls_measure = 'Result_by_sim_measure/Euclidean/kmeans_compute_k/clusteringMeasures/KMeansTransH'

#===========  KMeans find the optimal number of clusters  ================
#cls_measure = 'clusteringMeasures/KMeansTransH'
#cls_measure = 'clusteringMeasures/METISTransH'
```

```python
def load_cluster(cluster_addres, n_cls):
    cluster_list = []
    for i in range(n_cls):
        cls = cluster_addres + 'cluster-' + str(i) + '.txt'
        c_i = pd.read_csv(cls, delimiter=",", header=None)
        c_i.columns = ['o']
        c_i.replace('https://w3id.org/i40/sto#', '', regex=True, inplace=True)
        cluster_list.append(c_i)
    return cluster_list
```

```python
k=5
clusters_semep = []
for fold in range(k):
    cls_addres = cls_measure + str(fold) + '/clusters/'
    onlyfiles = [os.path.join(cls_addres, f) for f in listdir(cls_addres) if isfile(join(cls_addres, f))]
    num_cls = len(onlyfiles)
    clusters_semep.append(load_cluster(cls_addres, num_cls))
```

```python
def get_num_pair(n):
    return n*(n-1)/2


def intersection(lst1, lst2): 
    return set(lst1).intersection(lst2) 
```

```python
avg_category_score = []
for cls_fold in clusters_semep:
    category_score = []
    for cls in cls_fold:
        list_cls = list(cls['o'])
        if len(list_cls)==1:
            category_score.append(0)
            continue
        pair_sto = get_num_pair(len(list_cls))
        sum_partial = 0
        for ec in equivalence_class:
            # get the sto in ec and transform it in a list
            ec_str = str(ec).split('\n')[2:]
            ec = [cl.strip() for cl in ec_str]
            
            sto_intersected = intersection(list_cls, ec)
            pair_sto_intersected = get_num_pair(len(sto_intersected))
            sum_partial+=pair_sto_intersected/pair_sto
        category_score.append(sum_partial)#/len(equivalence_class))
    
    avg_category_score.append(sum(category_score) / len(category_score))


sum(avg_category_score) / len(avg_category_score)
```

```python
list_sto = ['a','b','c','d','f','g','rp','r']  #,'t','r'
equivalence_class = []
for sto in list_sto:
    a = same_equivalent_class(sto, X)
    flag = True
    for ec in equivalence_class:
        if (set(a).issubset(ec)):   #a <= equivalence_class
            flag = False
            break
        if set(ec).issubset(a):
            equivalence_class.remove(ec)
            flag = True
            break
    if flag:
        equivalence_class.append(a)
        flag = True
    print('equivalence_class: ', equivalence_class)
```
